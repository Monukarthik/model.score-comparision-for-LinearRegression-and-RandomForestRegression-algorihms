```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
dataset=pd.read_csv('Dataset.csv')
dataset.head()
```

```python
columns_contaminated=['Restaurant Name','City','Address','Locality','Locality Verbose']
for column in columns_contaminated:
  dataset[column]=dataset[column].str.replace('[^a-zA-Z0-9/s]','',regex=True)
dataset.columns
```

```python
dataset['Has Table booking']=dataset['Has Table booking'].replace({'Yes':1,'No':0})
dataset['Has Online delivery']=dataset['Has Online delivery'].replace({'Yes':1,'No':0})
dataset.head()
```

```python
dataset['Is delivering now']=dataset['Is delivering now'].replace({'Yes':1,'No':0})
dataset['Switch to order menu'].unique()
```

```python
dataset['Rating color'].unique()
```

```python
dataset['Rating color']=dataset['Rating color'].replace({'Dark Green':1, 'Green':2, 'Yellow':3, 'Orange':4, 'White':5, 'Red':6})
```

```python
dataset['Rating text'].unique()
```

```python
le=LabelEncoder()
dataset['Rating text']=le.fit_transform(dataset['Rating text'])
dataset
```

```python
dataset.columns
```

```python
x=dataset[['Country Code','Has Table booking','Has Online delivery','Is delivering now','Price range','Rating color','Rating text']].values
x
```

```python
y=dataset['Aggregate rating'].values
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

```python
predict=lr.predict(x_test)
```

```python
rf=RandomForestRegressor()
rf.fit(x_train,y_train)
```

```python
rf.score(x_train,y_train)
```
